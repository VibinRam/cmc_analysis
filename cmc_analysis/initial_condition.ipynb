```python
from cosmic.sample import InitialCMCTable

Singles, Binaries = InitialCMCTable.sampler(
    'cmc',
    binfrac_model=0.5,          # higher initial binaries
    primary_model='kroupa01',
    ecc_model='thermal',
    porb_model='log_uniform',
    qmin=0.1,
    cluster_profile='king',
    met=0.001,
    zsun=0.001,
    size=1000000,               # slightly larger
    w_0=7,                     # more concentrated
    virial_radius=0.5,         # slightly denser
    tidal_radius=50,           # realistic tidal field
    seed=1234
)

InitialCMCTable.write(Singles, Binaries, filename="cmc_test_vr0.5_bf0.5_zsun0.001_N1e6.hdf5")
```
