```python
def count_worldlines_with_multiple_ids(bh_worldlines):

    multi_id_worldlines = {
        wid: wl.id_history
        for wid, wl in bh_worldlines.items()
        if len(wl.id_history) > 1
    }

    return {
        "count": len(multi_id_worldlines),
        "wids": list(multi_id_worldlines.keys()),
        "id_histories": multi_id_worldlines,
    }

```

```python
def find_duplicate_ids_across_worldlines(bh_worldlines):
    """Check if any ID appears in the id_history of multiple worldlines."""
    
    id_to_wids = {}
    
    for wid, wl in bh_worldlines.items():
        for id_entry in wl.id_history:
            bh_id = id_entry['id']
            if bh_id not in id_to_wids:
                id_to_wids[bh_id] = []
            id_to_wids[bh_id].append(wid)
    
    duplicates = {bh_id: wids for bh_id, wids in id_to_wids.items() if len(wids) > 1}
    
    return {
        "total_ids": len(id_to_wids),
        "duplicate_ids": len(duplicates),
        "duplicates": duplicates,
    }
```
